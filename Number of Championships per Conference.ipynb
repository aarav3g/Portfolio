# Number of Championships Won by Each Conference (2000 - present)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame({
    'Michigan State': [1, 0, 0, 0, 0], #michigan state
    'Florida': [0, 2, 0, 0, 0], #florida 
    'Kentucky': [0, 1, 0, 0, 0], #kentucky
    'Duke': [0, 0, 3, 0, 0], #duke 
    'Maryland': [0, 0, 1, 0, 0], #maryland
    'UNC': [0, 0, 3, 0, 0], #unc
    'Virginia': [0, 0, 1, 0, 0], #virginia
    'Syracuse': [0, 0, 0, 1, 0], #syracuse
    'UConn': [0, 0, 0, 4, 0], #uconn 
    'Villanova': [0, 0, 0, 2, 0], #villanova
    'Louisville': [0, 0, 0, 1, 0], #louisville
    'Kansas': [0, 0, 0, 0, 2], #kansas
    'Baylor': [0, 0, 0, 0, 1], #baylor
    }, 
    index = ['Big10', 'SEC', 'ACC', 'BigEast', 'Big12'])


df.plot(kind = 'bar', stacked = True, figsize = (15,7), colormap = 'RdYlGn').legend(loc = 'upper left', ncol = 1) 
plt.title('Number of Championships Won by Each Conference (Modern)', fontsize = 15)
plt.xticks(rotation = 0, horizontalalignment = 'center')
plt.xlabel('Conference', fontsize = 12)
plt.ylabel('Number of Titles', fontsize = 12)
sns.set_style("whitegrid")
```

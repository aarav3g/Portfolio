# CBB Conference Championship Trends

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split

file_acc = 'acc championships.csv'
df_acc = pd.read_csv(file_acc)
file_sec = 'sec championships.csv'
df_sec = pd.read_csv(file_sec)
file_b1g = 'b1g championships.csv'
df_b1g = pd.read_csv(file_b1g)

x_acc = df_acc[['Year']]
y_acc = df_acc['PCT_ACC']
y_acc_c = df_acc['PCT_ACC_C']

x_sec = df_sec[['Year']]
y_sec = df_sec['PCT_SEC']
y_sec_c = df_sec['PCT_SEC_C']

x_b1g = df_b1g[['Year']]
y_b1g = df_b1g['PCT_B1G']
y_b1g_c = df_b1g['PCT_B1G_C']

x_train_acc, x_test_acc, y_train_acc, y_test_acc = train_test_split(x_acc, y_acc, train_size = 0.6, random_state=42)
x_train_acc_c, x_test_acc_c, y_train_acc_c, y_test_acc_c = train_test_split(x_acc, y_acc_c, train_size = 0.6, random_state=42)

x_train_sec, x_test_sec, y_train_sec, y_test_sec = train_test_split(x_sec, y_sec, train_size = 0.6, random_state=42)
x_train_sec_c, x_test_sec_c, y_train_sec_c, y_test_sec_c = train_test_split(x_sec, y_sec_c, train_size = 0.6, random_state=42)

x_train_b1g, x_test_b1g, y_train_b1g, y_test_b1g = train_test_split(x_b1g, y_b1g, train_size = 0.6, random_state= 42)
x_train_b1g_c, x_test_b1g_c, y_train_b1g_c, y_test_b1g_c = train_test_split(x_b1g, y_b1g_c, train_size = 0.6, random_state = 42)
```

## ACC Conference Championship Trends

```python
fig, ax = plt.subplots(figsize=(15,7))
k=1
model_acc = KNeighborsRegressor(n_neighbors=int(k))
model_acc.fit(x_train_acc, y_train_acc)
y_pred_acc = model_acc.predict(x_test_acc)

xvals_acc = np.linspace(x_acc.min(), x_acc.max(), 69)
ypreds_acc = model_acc.predict(xvals_acc)
ax.plot(xvals_acc, ypreds_acc * 100, '-', label = 'Overall', linewidth = 2, color = '#1b3ca4')

model_acc.fit(x_train_acc_c, y_train_acc_c)
y_pred_acc_c = model_acc.predict(x_test_acc_c)
ypreds_acc_c = model_acc.predict(xvals_acc)
ax.plot(xvals_acc, ypreds_acc_c * 100, '-', label = 'Conference', linewidth = 2, color = 'black')

ax.legend(loc = 'upper right', fontsize = 15)
ax.set_xlabel('Time', fontsize = 15)
ax.set_ylabel('Percent', fontsize = 15)
ax.set_title('Win Percent Needed to Win ACC Conference Championship Over Time', fontsize = 20)
plt.tight_layout()
```

## SEC Conference Championship Trends

```python
fig, ax = plt.subplots(figsize=(15,7))
k=1
model_sec = KNeighborsRegressor(n_neighbors=int(k))
model_sec.fit(x_train_sec, y_train_sec)
y_pred_sec = model_sec.predict(x_test_sec)

xvals_sec = np.linspace(x_sec.min(), x_sec.max(), 63)
ypreds_sec = model_sec.predict(xvals_sec)
ax.plot(xvals_sec, ypreds_sec * 100, '-', label = 'Overall', linewidth = 2, color = '#fac846')

model_sec.fit(x_train_sec_c, y_train_sec_c)
y_pred_sec_c = model_sec.predict(x_test_sec_c)
ypreds_sec_c = model_sec.predict(xvals_sec)
ax.plot(xvals_sec, ypreds_sec_c * 100, '-', label = 'Conference', linewidth = 2, color = 'black')

ax.legend(loc = 'lower right', fontsize = 15)
ax.set_xlabel('Time', fontsize = 15)
ax.set_ylabel('Percent', fontsize = 15)
ax.set_title('Win Percent Needed to Win SEC Conference Championship Over Time', fontsize = 20)
plt.tight_layout()
```
